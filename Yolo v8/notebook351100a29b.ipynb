```python
from ultralytics import YOLO

model = YOLO('yolov8n-cls.pt') 
```

```python
model.train(
    data='/kaggle/input/hath-food/food dataset/Images', 
    epochs=10, 
    imgsz=320,
    amp=False,
    device=0,       
    workers=4,      
    batch=64,       
    cache=True,

    fliplr=0.5,
    degrees=15,
    scale=0.5,
    erasing=0.4,
    hsv_s=0.4
)
```

```python
from IPython.display import Image, display


root_dir = '/kaggle/working/runs/classify/train4' 

display(Image(filename=f'{root_dir}/confusion_matrix.png'))

display(Image(filename=f'{root_dir}/results.png'))
```
